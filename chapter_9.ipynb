```python
# Shrink space around subplots to zero
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True) # sharex and sharey ensure that the plots share axes
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```

```python
# Change the drawstyle:
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data, color="black", linestyle="dashed", label="Default");
ax.plot(data, color="black", linestyle="dashed", drawstyle="steps-post", label="steps-post");
ax.legend();
```

```python
# Practice customizing axes:
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum());
# control where to set the x ticks
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
print("Type of ticks: {}".format(type(ticks)))
# control the axis labels
labels = ax.set_xticklabels(["one","two","three","four","five"], rotation=30, fontsize=8)
# Set x-label and title
ax.set_xlabel("Stages");
ax.set_title("A Groovy Plot in Stages");
```

```python
# Practice adding text to plots with an example of closing S&P index prices
from datetime import datetime
import pandas as pd

fig, ax = plt.subplots()

# Read from csv file:
data = pd.read_csv("datasets/spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]
print("Type of spx is: {}".format(type(spx)))

# Call plot function on the Series object spx:
spx.plot(ax=ax, color="black")

# Set important dates in subprime mortgage crisis:
crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]
print("Type of crisis data is: {}".format(type(crisis_data)))

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 76),
                xytext=(date, spx.asof(date) + 215),
                arrowprops=dict(facecolor="black", headwidth=4, width=1, headlength=2),
                horizontalalignment="left", verticalalignment="top");
    
# Set axis limits and title:
ax.set_xlim(["1/1/2007", "1/1/2011"]);
ax.set_ylim([600, 1800]);
ax.set_title("Important dates in the subprime mortgage crisis");
```

```python
# Draw simple shapes
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.7, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.3), 0.2, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color="green", alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
