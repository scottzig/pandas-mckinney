```python
# Practice adding text to plots with an example of closing S&P index prices
from datetime import datetime
import pandas as pd

fig, ax = plt.subplots()

# Read from csv file:
data = pd.read_csv("datasets/spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]
print("Type of spx is: {}".format(type(spx)))

# Call plot function on the Series object spx:
spx.plot(ax=ax, color="black")

# Set important dates in subprime mortgage crisis:
crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]
print("Type of crisis data is: {}".format(type(crisis_data)))

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 76),
                xytext=(date, spx.asof(date) + 215),
                arrowprops=dict(facecolor="black", headwidth=4, width=1, headlength=2),
                horizontalalignment="left", verticalalignment="top");
    
# Set axis limits and title:
ax.set_xlim(["1/1/2007", "1/1/2011"]);
ax.set_ylim([600, 1800]);
ax.set_title("Important dates in the subprime mortgage crisis");
```

```python
# Draw simple shapes
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.7, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.3), 0.2, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color="green", alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
# Example with restaurant tips

tips = pd.read_csv("datasets/tips.csv")
print("Type of tips: {}".format(type(tips)))
tips.head()
```

```python
# Extract just the day and party size
party_counts = pd.crosstab(tips["day"], tips["size"]) # crosstab computes a frequency table from two DF columns
party_counts = party_counts.reindex(index = ["Thur", "Fri", "Sat", "Sun"])
party_counts
```

```python
# Get rid of one and six person parties
party_counts = party_counts.loc[:, 2:5]
# Normalize each row to sum to 1
party_pcts = party_counts.div(party_counts.sum(axis="columns"), axis="index")
party_pcts
```

```python
# Make a stacked barplot of the percentages
party_pcts.plot.bar(stacked=True);
```

```python
# Use seaborn to compute tipping percentage by day
import seaborn as sns
# Make new column holding the tip percentage
tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
tips.head()
```

```python
# Use seaborn's barplot function
sns.barplot(x="tip_pct", y="day", data=tips, orient="h");
```

```python
# Split by additional category (here time of day)
sns.barplot(x="tip_pct", y="day", hue="time", data=tips, orient="h");
```

```python
# Make a histogram of tip percentages of the total bill
tips["tip_pct"].plot.hist(bins=200);
```

```python
# Make density plot using a mixture of Gaussians
tips["tip_pct"].plot.density();
```

```python
# Make histogram of bimodal Gaussian using seaborn
comp1 = np.random.standard_normal(200)
comp2 = 10 + 2*np.random.standard_normal(200)
values = pd.Series(np.concatenate([comp1, comp2]))

# Plot histogram and density plot
fig, axes = plt.subplots(2, 1, sharex=True)
sns.histplot(values, bins=100, color="black", ax=axes[0]);
sns.kdeplot(values, ax=axes[1]);
```

```python
# Do some macrodata refinement
macro = pd.read_csv("datasets/macrodata.csv")
data = macro[["cpi", "m1", "tbilrate", "unemp"]]
data
```

```python
# Take log of data, drop any na values
trans_data = np.log(data).diff().dropna()
trans_data.tail()
```

```python
# Make a scatter plot and fit a line to the data
ax = sns.regplot(x="m1", y="unemp", data=trans_data)
ax.set_title("Changes in log(m1) versus log(unemp)");
```

```python
# Make pair plots from the macrodata
sns.pairplot(trans_data, diag_kind="kde", plot_kws={"alpha": 0.5});
```

```python
# Make facet grid split by day of the week, time of meal, and smoking status
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker", kind="bar", data=tips[tips.tip_pct<1]);
```

```python
# Another facet grid with an additional row for time of meal
sns.catplot(x="day", y="tip_pct", row="time", col="smoker", kind="bar", data=tips[tips.tip_pct<1]);
```

```python
# Make a box and whisker plot with seaborn
sns.catplot(x="tip_pct", y="day", kind="box", data=tips[tips.tip_pct<0.5])
```
